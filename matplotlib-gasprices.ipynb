
### Creating a data frame

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
lpg = pd.read_csv('lpg gas.csv')
lpg
```

### Plotting a line graph for domestic RSP

```python
plt.plot(lpg['Date Effective From'] , lpg['RSP of Domestic LPG (Rs./14.2Kg Cylinder)'])
plt.show()
```

### Plotting a line graph for commercial RSP

```python
plt.plot(lpg['Date Effective From'] , lpg['RSP of Commercial LPG (Rs./19Kg Cylinder)'])
plt.show()
```

### Plotting a line graph for both domestic and commercial

```python
plt.plot(lpg['Date Effective From'] , lpg['RSP of Domestic LPG (Rs./14.2Kg Cylinder)'])
plt.plot(lpg['Date Effective From'] , lpg['RSP of Commercial LPG (Rs./19Kg Cylinder)'])
plt.show()
```

### Percentage increase calculation
      

```python
lpg['percentage_increase for Domestic'] = lpg['RSP of Domestic LPG (Rs./14.2Kg Cylinder)'].pct_change() * 100
lpg
```

```python
lpg['percentage_increase for Commercial'] = lpg['RSP of Commercial LPG (Rs./19Kg Cylinder)'].pct_change() * 100
lpg
```
